Visualizing Twitch Network of Followers
Who follows who? What categories overlap? 

1. Scrape top 1000 streams (viewers vs follows)
    - filter english only streams
2. Get followers for each stream
3. Network visualization
    - nodes represent streamers
    - group nodes by category
    - node size = follower size
    - edges represent overlapping followers (how can we visualize more followers? )

```python
from itertools import islice
from twitch import TwitchHelix
```

```python
stream_array = []
stream_iterator = client.get_streams(page_size=100)

for stream in islice(stream_iterator, 0, 1000):
    stream_array.append(stream)
```

```python
stream_array[0]['user_name']

stream_dict = {'user_name' : [],
               'user_id' : []}

for i in range(len(stream_array)):
    if stream_array[i]['language'] == 'en':
        stream_dict['user_name'].append(stream_array[i]['user_name'])
        stream_dict['user_id'].append(stream_array[i]['user_id'])
```
